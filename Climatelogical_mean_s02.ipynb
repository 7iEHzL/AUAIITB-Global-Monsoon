```python
path1 = "C:/Users/user/Downloads/Met_data/pcp_S_Korea_1970_daily.nc"
path2 = "C:/Users/user/Downloads/Met_data/pcp_S_Korea_2020_daily.nc"
path3 = "C:/Users/user/Downloads/Met_data/T_RH_q_S_Korea_1970_daily.nc"
path4 = "C:/Users/user/Downloads/Met_data/T_RH_q_S_Korea_2020_daily.nc"
country = "South_Korea"
data1 = xr.open_dataset(path1)
data2 = xr.open_dataset(path2)
data3 = xr.open_dataset(path3)
data4 = xr.open_dataset(path4)
```

```python
ds = nc.Dataset(path1)
#print(ds.variables)
time = ds.variables['time']
lat = ds.variables['latitude']
lon = ds.variables['longitude']
precip = ds.variables['tp']

ds2 = nc.Dataset(path2)
#print(ds2.variables)
time2 = ds2.variables['time']
lat2 = ds2.variables['latitude']
lon2 = ds2.variables['longitude']
precip2 = ds2.variables['tp']
```

```python
daily_spatial_average_precip = np.mean(precip,axis=(1,2))*1000
daily_spatial_average_precip2 = np.mean(precip2,axis=(1,2))*1000
```

```python
plt.figure(figsize=(10,6))
plt.plot(time[:],daily_spatial_average_precip,label='Daily average prcp_1970')
plt.plot(time2[:],daily_spatial_average_precip2,label='Daily average prcp_2020')
plt.xlabel("Months")
plt.ylabel("precip")

custom_dates = [1,30,60,90,121,151,182,213,243,274,304,336]
custom_labels = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

plt.xticks(custom_dates,custom_labels)
plt.title('Daily spatial avg.prcp.')
plt.legend()
plt.grid(True)
#plt.savefig('average_precipitation_1970')

plt.show()
```

```python
ds3 = nc.Dataset(path3)
#print(ds.variables)
time3 = ds3.variables['time']
lat3 = ds3.variables['latitude']
lon3 = ds3.variables['longitude']
R_H3 = ds3.variables['r'][:,0,:,:]

ds4 = nc.Dataset(path4)
#print(ds.variables)
time4 = ds4.variables['time']
lat4 = ds4.variables['latitude']
lon4 = ds4.variables['longitude']
R_H4 = ds4.variables['r'][:,0,:,:]

daily_spatial_average_R_H3 = np.mean(R_H3,axis=(1,2))
daily_spatial_average_R_H4 = np.mean(R_H4,axis=(1,2))
plt.figure(figsize=(10,6))
plt.plot(time3[:],daily_spatial_average_R_H3,label='Daily average R_H_1970')
plt.plot(time4[:],daily_spatial_average_R_H4,label='Daily average R_H_2020')
plt.xlabel("Months")
plt.ylabel("R_H(%)")

custom_dates = [1,30,60,90,121,151,182,213,243,274,304,336]
custom_labels = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

plt.xticks(custom_dates,custom_labels)
plt.title('Daily spatial avg.R_H.')
plt.legend()
plt.grid(True)
#plt.savefig('average_precipitation_1970')

plt.show()
```

```python
ds3 = nc.Dataset(path3)
#print(ds.variables)
time3 = ds3.variables['time']
lat3 = ds3.variables['latitude']
lon3 = ds3.variables['longitude']
q3 = ds3.variables['q'][:,0,:,:]

ds4 = nc.Dataset(path4)
#print(ds.variables)
time4 = ds4.variables['time']
lat4 = ds4.variables['latitude']
lon4 = ds4.variables['longitude']
q4 = ds4.variables['q'][:,0,:,:]

daily_spatial_average_q3 = np.mean(q3,axis=(1,2))
daily_spatial_average_q4 = np.mean(q4,axis=(1,2))
plt.figure(figsize=(10,6))
plt.plot(time3[:],daily_spatial_average_q3,label='Daily average q_1970')
plt.plot(time4[:],daily_spatial_average_q4,label='Daily average q_2020')
plt.xlabel("Months")
plt.ylabel("q(kg/kg)")

custom_dates = [1,30,60,90,121,151,182,213,243,274,304,336]
custom_labels = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

plt.xticks(custom_dates,custom_labels)
plt.title('Daily spatial avg.q.')
plt.legend()
plt.grid(True)
#plt.savefig('average_precipitation_1970')

plt.show()
```

```python
from scipy.stats import linregress
from matplotlib.pyplot import Line2D
slope1,intercept1,r_value1,p_value1,std_err1 = linregress(daily_spatial_average_temp3,daily_spatial_average_q3)
slope2,intercept2,r_value2,p_value2,std_err2 = linregress(daily_spatial_average_temp4,daily_spatial_average_q4)
plt.figure(figsize=(10,6))
plt.scatter(daily_spatial_average_temp3,daily_spatial_average_q3,label='1970_data')
plt.plot(daily_spatial_average_temp3,slope1*daily_spatial_average_temp3+intercept1)
plt.scatter(daily_spatial_average_temp4,daily_spatial_average_q4,label='2020_data')
plt.plot(daily_spatial_average_temp4,slope2*daily_spatial_average_temp4+intercept2)
plt.xlabel("Temperature(K)")
plt.ylabel("q kg/kg)")
plt.title("avg temp(K) vs avg q(kg/kg) in 1970 and 2020 South_Korea")
legend_text = (f"Slope1: {slope1:.5f}\n"
               f"r_value1: {r_value1:.5f}\n"
               f"std_err1: {std_err1:.5f}\n"
               f"p-value1: {p_value1:.5f}\n"
               f"Slope2: {slope2:.5f}\n"
               f"r_value2: {r_value2:.5f}\n"
               f"std_err2: {std_err2:.5f}\n"
               f"p-value2: {p_value2:.5f}\n")

plt.legend(title=legend_text)
plt.grid(True)
plt.show()


print(slope1,slope2)
print(p_value1,p_value2)
```
